```python
import numpy as np
from scapy.utils import rdpcap

# When scapy is used as a library, rdpcap requires to import the right layers - if not, it falls back to raw inputs with pcapreader
# this is a different behavior from the CLI
import scapy.layers.l2
import scapy.layers.inet
```

```python
# reads a full *.pcap file and loads it into memory
# beware of memory use
# pcapreader is an alternative based on an iterator

example_file_path = "/home/benjamin/Folders_Python/Cyber/data/home.pcap"

p = rdpcap(example_file_path)

# the output is a PacketList object

type(p)
```

```python
# list the packets in the PacketList object
print(f"Il y a {len(p)} paquets")
p.nsummary()
```

```python
# the .sessions() method assembles the unidirectional streams
# into a dictionnary

streams = p.sessions()

print(type(streams))
```

```python
print(f"Il y a {len(streams)} streams")
for k,v in streams.items():
    print(k, "---", v)
```

```python
# zoom in one of the streams, chosen at random
# each stream is a PacketList

streams_keys = list(streams.keys())

i = np.random.randint(len(streams_keys))

stream_key = streams_keys[i]
stream_content = streams.get(stream_key)

print(f"stream numéro {i} / {len(streams_keys)}")
stream_content
```

```python
# listing the packets in the stream
for i, pkt in enumerate(stream_content):
    print(f"packet {i} / {len(stream_content)-1}")
    print(pkt.show())
    print("\n")
```
